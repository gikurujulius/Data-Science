```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
ice = pd.read_csv('SeaIce.txt', delim_whitespace = True)
ice
```

```python
#On inspecting the data further, there are some errors -999 on the data type field. 
#CLeaning
clean_ice_data = ice[ice.data_type != '-9999']
```

```python
#Visualizing the relationship between the year and the sea ice extent
sns.implot('year', 'extent', clean_ice_data)
```
